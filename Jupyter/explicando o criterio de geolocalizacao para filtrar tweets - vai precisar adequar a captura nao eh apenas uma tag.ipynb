```python
import tweepy as tw
import pandas as pd
'''
 é necessário se inscrever como desenvolvedor no twitter par obter as chaves de acesso abaixo. o processo é rápido e simples
 basta ter uma conta twitter
 
 esse método eu estudei por um pedido do Leon... o problema é que o Twitter inibiu o geocode , então a captura atual 
 que eu estava fazendo, com os módulos "Exporter e TweetManager" terão que ser adaptadas ...eu imaginei que era só por a 
 tag geocode lá, mas não funciona mais(desde meados do ano passado...).  Então fui por outro caminho e descobri essa alternativa.  
 
 É possível vc capturar o twitter dada uma coordenada GPS...e traçar um raio de ação. Por exemplo, pega a rua conde de 
 bonfim a altura da saens pena , no Google Maps, vai dar uma coordenada ""-22.929956, -43.237484", aí vc escolhe um raio de 3km. e poe 
 uma tag de busca, então por esse comando abaixo, vc teria os tweets com aquela tag de busca, naquela região...
 
 ai eu trago os dados da mensagem, a data, o id-tweet, e o local da mensagem...
 bom, até aqui, foi o que consegui...
 
 vou testar isso melhor... o que imagino é adequar essa nova forma de capturar e colocar o Tweetmanager aqui dentro, é uma
 possibilidade...
 ou seja, colocar, no export, essa inteligência da georeferencia...  
 o meu receio é que os twitters antigos não tem vindo... não sei se é devido ao covid, quarentena, haver muito tráfego...ou]
 se eles restringiram mesmo... nesse caso, vamos ter que ver alternativa para pegar os tweets do início da pandemia..deve ter
 bases disponíveis...
'''
consumer_key = 'xpto'
consumer_secret = 'xpto'
access_token = 'xpto'"
access_token_scret = 'xpto'
auth = tw.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_scret)

api = tw.API(auth)

public_tweets = api.home_timeline()
busca = "coronavirus" + "-filter:retweets"
'''
Método que faz a busca dos tweets, no geocode, coloquei um endereço aqui do flamengo, e escolhi um raio de 5km, o tweet_code, é o tipo de retono, coloquei para buscar todas as informações possiveis q=busca, é o argumento de pesquisa esta procurando que está na celula de cima. e em .items(10) ele busca os 100 primeiros tweets na ordem pela data do twitter se não colocar uma quantidade, ele vai buscar todos os tweets com esse determinado assunto, pode travar o notebook.  A coordenada do geocode pode ser obtido no Google Maps, clicando na rua...
'''

tweets = tw.Cursor(api.search,q=busca, geocode="-22.936922, -43.175749,5km",tweet_mode='extended').items(100)

'''

Crio um dicionário que vai armazenar todos os tweets capturados para uma futura exportação 
'''
dados = []

'''
Aqui ele percorre todos os tweets capturados e seleciona só as informações que eu quiser, aqui eu coloquei algunas informações, deixei um print para poder ver o que ele recupera

'''
for tweet in tweets:
  twitter = tweet._json
  print(twitter)
  dataCriacao = ''
  texto = ''
  idTwitter = ''
  lugar_marcado = '-'
  nome = ''
  nomeUsuario = ''
  lugar_registro_usuario = '-'
  dataCriacao = twitter['created_at']
  texto = twitter['full_text']
  idTwitter = twitter['id_str']
  if twitter['place'] is None:
    lugar_marcado = '-'
  else:
    lugar_marcado = twitter['place']['full_name']
  nome = twitter['user']['name']
  nomeUsuario = twitter['user']['screen_name']
  if 'media' in twitter['entities']:
    link = twitter['entities']['media'][0]['expanded_url']
  lugar_registro_usuario = twitter['user']['location']
  linha = [dataCriacao,texto,idTwitter,lugar_marcado,nome,nomeUsuario,link,lugar_registro_usuario]
  dados.append(linha)
  '''
  

Aqui ocorre a  exportação seja para csv quanto para excel . tem um icone parecendo uma pasta, onde que ele gera os arquivos , caso não apareça o arquivo depois de rodar todas as celular, clique em refresh !

  '''
  
  
arquivo = pd.DataFrame(dados,columns=['Data de Criação','Texto','Id do Twitter','Lugar Mencionado','Nome','Nome do Usuário', 'Link','Lugar do Registro do Usuário'])arquivo.to_csv("captura de twitter.csv",sep=';')
arquivo.to_excel("captura de twitter.xls")
arquivo.head()
  
```
